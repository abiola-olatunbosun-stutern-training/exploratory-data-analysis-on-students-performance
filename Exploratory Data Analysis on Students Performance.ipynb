### 4. Create a new variable average_score represented from math, reading and writing scores.

```python
# Creating new column average_score in dataframe by calculating average of math score, reading score, writing score, and rounded up result to 2 float/decimal points 
students_performance["average_score"] = round((students_performance["math score"] + students_performance["reading score"] + students_performance["writing score"])/3, 2)

# Display first five rows in dataframe
students_performance.head()
```

### 5. Create another variable average_score_cat that categorises the average_score using WAEC grading system.

```python
# Defining a function to categorize average_score using WAEC grading system
def average_score_cat(average_score):
        if average_score >= 85 and average_score <= 100:  # average_scores from 85 to 100 categorized as A1
            return "A1"
        elif average_score >= 70 and average_score < 85:  # average_scores from 70 to 84 categorized as B2
            return "B2"
        elif average_score >= 65 and average_score < 70:  # average_scores from 65 to 69 categorized as B3
            return "B3"
        elif average_score >= 60 and average_score < 65:  # average_scores from 60 to 64 categorized as C4
            return "C4"
        elif average_score >= 55 and average_score < 60:  # average_scores from 55 to 59 categorized as C5
            return "C5"
        elif average_score >= 50 and average_score < 55:  # average_scores from 50 to 54 categorized as C6
            return "C6"
        elif average_score >= 45 and average_score < 50:  # average_scores from 45 to 49 categorized as D7
            return "D7"
        elif average_score >= 40 and average_score < 45:  # average_scores from 40 to 44 categorized as E8
            return "E8"
        else:
            return "F9"                                   # average_scores below 40 categorized as F9


# Creating new column average_score_cat in dataframe by applying above fucntion to each element of column average_score
students_performance["average_score_cat"] = students_performance["average_score"].apply(lambda average_score: average_score_cat(average_score))

# Display first five rows of dataframe
students_performance.head()
```

### 6. Find a relationship (if any) between each variable and the new variable average_score_cat.

```python
# Finding the relationship between variables using correlation function, rounded to 2 float/decimal points
round(students_performance.corr(), 2)
```
